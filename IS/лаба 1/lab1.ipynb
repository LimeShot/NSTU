```python
#  ""/"-"   ,   -   
tressholdDict = {
    'SelectByHotkeys' : (0, 0.0105),
    'ActionLatency': (24, 108),
    'NumberOfPACs': (0.0008, 0.0059),
    'AssignToHotkeys': (0, 0.0009375),
    'GapBetweenPACs': (6.25, 77),
    'APM': (20, 230)
}

print (tressholdDict)
```

```python
for feature_name  in tressholdDict:
    threesMin =  tressholdDict[feature_name][0] #   
    tresshMax =  tressholdDict[feature_name][1] #   

    #  - .   .  - 
    over_count = len(  df_digital[
                            ~((df_digital[feature_name] >= threesMin ) & (df_digital[feature_name] <= tresshMax)) #    . ~   
                          ]
                    )

    dcount = df_digital[feature_name].count() # - . 

    print ('''{}.     {}.
  - .   - {}  {} () ({:.1%})\n'''.format(
        feature_name,
        tressholdDict[feature_name],
        over_count,
        dcount,
        over_count/ dcount,
    ))
```

```python
#          
df_digPlusFiltred = df_digital.copy()

for feature_name  in tressholdDict:
    threesMin =  tressholdDict[feature_name][0]
    tresshMax =  tressholdDict[feature_name][1]


    #    .  "~" -  
    df_digPlusFiltred = df_digPlusFiltred[
        (df_digPlusFiltred[feature_name] >= threesMin ) & (df_digPlusFiltred[feature_name] <= tresshMax)
        ]
    print(" ", feature_name )


print (" {} .  {} .".format(
    len(df_digital)  - len(df_digPlusFiltred),
    len(df_digPlusFiltred))
)
```

```python
#     :
# data[features].hist() #  

plt.figure(figsize=(16, 5))
plot_number = 0

for feature_name in (features+target):
    plot_number += 1

    plt.subplot(1, len(features+target), plot_number)

    plt.hist(df_digPlusFiltred[feature_name]) #  

    plt.title(feature_name)
    plt.xlabel(u'')
    plt.ylabel(u'')
    print (feature_name,
           df_digPlusFiltred[feature_name].min(),
           df_digPlusFiltred[feature_name].max())
```

```python
df_digPlusFiltred.to_csv('./datasets/melbourne_data_PREPARED.csv', #   
            index=False, #   "" (.  )
           )
```

```python
from sklearn.model_selection import train_test_split
valid_size = 0.3 #     
rand_seed = 8 #    . 

x_train, x_test, y_train, y_test =  train_test_split(
    df_digPlusFiltred[features],    #   X
    df_digPlusFiltred[target],      #   y

    test_size = valid_size, #     
    random_state=rand_seed, #    . 
    shuffle=True            # 
)
```

```python
print ("- : \n  x_train: {}, y_train {} \n  x_test:  {}, y_test  {} \n  total x: {}, total y {} ".format  (
    len(x_train), len(y_train),
    len(x_test),  len(y_test),
    len(x_train)+len(x_test), len(y_train)+len(y_test),
))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_ActionLatency = LinearRegression()
```

```python
model_ActionLatency.fit(
    x_train[['ActionLatency']],
    y_train[['APM']]
)
```

```python
y_predicted_on_ActionLatency = model_ActionLatency.predict(x_test[['ActionLatency']])
```

```python
#       
plt.figure(figsize=(15,5))

#   ( ) - plt.scatter
plt.scatter(x_train[['ActionLatency']],  y_train[['APM']], alpha=0.3, label='') # 
plt.scatter(x_test[['ActionLatency']],   y_test[['APM']],  alpha=0.3, label='')   # 

#   - ,    - plt.plot
plt.plot( x_test[['ActionLatency']], y_predicted_on_ActionLatency,
          lw=2, color='red',
          label="y_pred (X) = x0 * {:.3f} + {:.3f}".format(model_ActionLatency.coef_[0][0], model_ActionLatency.intercept_[0]))
plt.grid(True)  # .   color='black', linewidth=0.7
plt.legend(loc="best")  #  
plt.xlabel('ActionLatency') #  
plt.ylabel(target[0])   #  
plt.title(u' '); # 
```

```python
plt.figure(figsize=(6,4))

#     - sns.distplot
# data = y_true - y_pred
sns.histplot( data = y_test[['APM']] - y_predicted_on_ActionLatency,
              color="red",
              kde=True, #     
              stat="density",# density:    1
            )

#  
plt.plot(
    [0, 0],
    [0, 10 * pow(10,-7)], '--', lw=2, c='r')

plt.ylabel(u'')
plt.xlabel(u' ')
plt.title(u'    ');
plt.show()
```

```python
def PLOT34(y_test, y_pred) -> None:
    '''
      
    :param y_test: -    
    :param y_pred: -    
    '''
    plt.figure(figsize=(12,6))

    # ==     ==
    #    121: 1-, 2-, 1-    
    plt.subplot(121)
    plt.scatter(y_test, y_pred,  alpha=0.1, color = "#17becf")
    plt.plot(  [ np.min(y_test), np.max(y_test) ], # x1,x2
               [ np.min(y_test), np.max(y_test) ], # y1,y2
               '--',
               alpha=0.7, lw=3, color = "black")
    plt.title('   ');
    plt.xlabel(' Y')
    plt.ylabel(' Y')
    plt.grid(True)  # .   color='black', linewidth=0.7

    # ==    ==
    #    121: 1-, 2-, 2-    
    plt.subplot(122)
    plt.scatter(y_test, (y_test - y_pred)**2,  alpha=0.1, color = "#17becf")
    plt.title('    ')
    plt.xlabel(' Y')
    plt.ylabel('  ')
    plt.grid(True)  # .   color='black', linewidth=0.7
#--------------------------------------------------------------------------

PLOT34(
    y_test = y_test[['APM']],
    y_pred = y_predicted_on_ActionLatency
)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print ("  .  'ActionLatency'.")
print ("   y(X) = k0*x0 + ... kN*xN + b")
print(f"y = f(X) = ")
for i, kN in enumerate (model_ActionLatency.coef_):
    print(f" {' ' if i==0 else '+'}  {kN[0]:.3f} * x{i} ")
print(f"+ {model_ActionLatency.intercept_[0]:.3f} ")
print("\n")

'''
   : MAE, RMSE, R2.
     .
'''
mae = mean_absolute_error(y_test, y_predicted_on_ActionLatency)
mse = mean_squared_error(y_test, y_predicted_on_ActionLatency)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_predicted_on_ActionLatency)
print ("MAE : {:>9,.3f} (  )".format( mae ))
print ("MSE : {:>9,.6f} ( )".format( mse ))
print ("RMSE: {:>9,.6f} (.    )".format( rmse ))
print ("R2  : {:>9,.3f} (. )".format( r2 )) #,       ( 0  1)
#   60%   .  40%          ,     
#  ,  R2 > 0.7,    R2  0.5,   0.5          
```

```python
def PLOT34(y_test, y_pred) -> None:
    '''
      
    :param y_test: -    
    :param y_pred: -    
    '''
    plt.figure(figsize=(12,6))

    # ==     ==
    #    121: 1-, 2-, 1-    
    plt.subplot(121)
    plt.scatter(y_test, y_pred,  alpha=0.1, color = "#17becf")
    plt.plot(  [ np.min(y_test), np.max(y_test) ], # x1,x2
               [ np.min(y_test), np.max(y_test) ], # y1,y2
               '--',
               alpha=0.7, lw=3, color = "black")
    plt.title('   ');
    plt.xlabel(' Y')
    plt.ylabel(' Y')
    plt.grid(True)  # .   color='black', linewidth=0.7

    # ==    ==
    #    121: 1-, 2-, 2-    
    plt.subplot(122)
    plt.scatter(y_test, (y_test - y_pred)**2,  alpha=0.1, color = "#17becf")
    plt.title('    ')
    plt.xlabel(' Y')
    plt.ylabel('  ')
    plt.grid(True)  # .   color='black', linewidth=0.7
#--------------------------------------------------------------------------

PLOT34(
    y_test = y_test[['APM']],
    y_pred = y_predicted_on_OtherFeatures
)
```
