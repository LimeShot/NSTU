```python
import numpy as np
```

```python
np.random.seed(42)

def y(x):
    return 3*x - 3

n = 3
x = np.random.uniform(low = 3, high = 20, size = n)
x = np.sort(x)
z = y(x)

```

```python
print(np.sort(x))
```

```python
print(np.sort(z))
```

```python
w = np.random.normal(0, 1, size = n)
z = z + w
```

```python
print(w)
```

```python
print(np.sort(z))
```

```python
def regressor(x):
   return np.array((1, x))

A  = []
for xi in x:
   A.append(regressor(xi))
A = np.array(A)

a = np.linalg.inv(A.T@A)@A.T@z
```

```python
a
```

```python
from matplotlib import pyplot as plt

xmin = np.min(x)
xmax = np.max(x)

plt.scatter(x, z, color = 'red')
plt.plot([xmin, xmax], [a[0]+ xmin*a[1], a[0]+ xmax*a[1]], label = 'Оценка')
plt.plot([xmin, xmax], [y(xmin), y(xmax)], label = 'Истинная функция', linestyle = '--')
plt.legend()
```

```python
def y(x):
    return -2*x**2 + 3*x - 3

n = 4
x = np.random.uniform(low = -5, high = 5, size = n)
x = np.sort(x)
z = y(x)
w = np.random.normal(0, 1, len(z))
z = z + w
```

```python
x
```

```python
z
```

```python
def regressor(x):
   return np.array((1, x, x**2))

A  = []
for xi in x:
   A.append(regressor(xi))
A = np.array(A)

a = np.linalg.inv(A.T@A)@A.T@z
```

```python
a
```

```python
xmin = np.min(x)
xmax = np.max(x)

xlin = np.linspace(xmin, xmax, 100)

regression = []
for x_loc in xlin:
    regress = a[0] + a[1]*x_loc + a[2]*x_loc**2
    regression.append(regress)

plt.scatter(x, z, color = 'red')
plt.plot(xlin, regression, label = 'Полученная оценка')
plt.plot(xlin, y(xlin), label = 'Истинная функция', linestyle = '--')
plt.legend()
```

```python


x = np.linspace(-3, 3, 7)
y = np.random.uniform(-3, 10, size = len(x))

plt.scatter(x, y)
```

```python
display(x, y)
```

```python
xmin = np.min(x)
xmax = np.max(x)

xlin = np.linspace(xmin, xmax, 100)
```

```python
def regressor(x):
   return np.array((1, x, x**2))

A  = []
for xi in x:
   A.append(regressor(xi))
A = np.array(A)

a = np.linalg.inv(A.T@A)@A.T@y

regression = []
for x_loc in xlin:
    regress = 0
    for i in range(len(a)):
        regress += a[i]*x_loc**i
    regression.append(regress)
    
    
plt.scatter(x, y, color = 'red')
plt.plot(xlin, regression, label = 'Полином 2-ой степени')
plt.legend()

RSS = 0

for j, xi in enumerate(x):
    yi = a[0]
    for i in range(1, len(a)):
        yi += a[i]*xi**i
    RSS += (yi - y[j])**2
print(RSS)
print(a)
    
```
