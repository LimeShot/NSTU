```python
import numpy as np
```

```python
def y(x):
    return 3*x - 3

n = 3
x = np.random.uniform(low = 3, high = 20, size = n)
z = y(x)
#e = np.random.normal(0, 10, len(z))

e = np.random.uniform(-10, 10, size = n)

z = z + e
```

```python
def regressor(x):
   return np.array((1, x))

X  = []
for xi in x:
   X.append(regressor(xi))
X = np.array(X)

Theta = np.linalg.inv(X.T@X)@X.T@z
```

```python
Theta
```

```python
from matplotlib import pyplot as plt

xmin = np.min(x)
xmax = np.max(x)

plt.scatter(x, z, color = 'red')
plt.plot([xmin, xmax], [Theta[0]+ xmin*Theta[1], Theta[0]+ xmax*Theta[1]], label = 'regression')
plt.plot([xmin, xmax], [y(xmin), y(xmax)], label = 'true function', linestyle = '--')
plt.legend()
```

```python
def y(x):
    return -2*x**2 + 3*x - 3

n = 4
x = np.random.uniform(low = -5, high = 5, size = n)
z = y(x)
e = np.random.normal(0, 1, len(z))
z = z + e
```

```python
def regressor(x):
   return np.array((1, x, x**2))

X  = []
for xi in x:
   X.append(regressor(xi))
X = np.array(X)

Theta = np.linalg.inv(X.T@X)@X.T@z
```

```python
Theta
```

```python
xmin = np.min(x)
xmax = np.max(x)

xlin = np.linspace(xmin, xmax, 100)

regression = []
for x_loc in xlin:
    regress = Theta[0] + Theta[1]*x_loc + Theta[2]*x_loc**2
    regression.append(regress)

plt.scatter(x, z, color = 'red')
plt.plot(xlin, regression, label = 'regression')
plt.plot(xlin, y(xlin), label = 'true function', linestyle = '--')
plt.legend()
```

```python
np.random.seed(42)

x = np.linspace(-3, 3, 7)
y = np.random.uniform(-4, 10, size = len(x))

plt.scatter(x, y)
```

```python
display(x, y)
```

```python
xmin = np.min(x)
xmax = np.max(x)

xlin = np.linspace(xmin, xmax, 100)
```

```python
def regressor(x):
   return np.array((1, x, x**2))

X  = []
for xi in x:
   X.append(regressor(xi))
X = np.array(X)

Theta = np.linalg.inv(X.T@X)@X.T@y

regression = []
for x_loc in xlin:
    regress = 0
    for i in range(len(Theta)):
        regress += Theta[i]*x_loc**i
    regression.append(regress)
    
    
plt.scatter(x, y, color = 'red')
plt.plot(xlin, regression, label = 'regression')
plt.legend()

RSS = 0

for j, xi in enumerate(x):
    yi = Theta[0]
    for i in range(1, len(Theta)):
        yi += Theta[i]*xi**i
    RSS += (yi - y[j])**2
print(RSS)
    
```
